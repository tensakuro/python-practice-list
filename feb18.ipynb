```python
def count_chars(str1):

    checked = ""

    for i in str1:

        # skip already counted character
        if i in checked:
            continue

        cnt = 0

        for j in str1:
            if i == j:
                cnt += 1

        print(i, ":", cnt)

        # mark character as processed (ONLY ONCE)
        checked += i


count_chars("BANANA")
```

```python
# nested loops 

def common_char(s1,s2):
    visited=""
    for i in s1:
        if i in visited:
            continue

        for j in s2:
            if i==j:
                print(i,end=" ")
                visited+=i
                break

common_char("banana",'animal')
```
