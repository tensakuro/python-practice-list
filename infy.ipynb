```python
def two_sum_bruteforce(nums, target):
    n = len(nums)
    
    for i in range(n):
        for j in range(i+1, n):
            if nums[i] + nums[j] == target:
                return [i,j]
            
nums = [10, 15, 3, 7]
target = 17


two_sum_bruteforce(nums,target)
            
```
