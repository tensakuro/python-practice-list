```python
# check if a character is a letter

import string
def is_letter(ch):
    return ch in string.ascii_letters

is_letter('A')
is_letter('9')
```

```python
# check if a character is a digit

def is_digit(ch):
    return ch in string.digits

is_digit('6')
```

```python
def count_digits(s):
    return sum(1 for ch in s if ch in string.digits)

count_digits("abc123")
```

```python
# remove punctuation
def remove_punctuation(s):
    return ''.join(ch for ch in s if ch not in string.punctuation)

remove_punctuation("hello!!! how, are you??")

```

```python
# def remove_digits


def remove_digits(s):
    return ''.join(ch for ch in s if ch not in string.digits)

remove_digits("abc123xyz")
```

```python
print("ab\tc\nd")

def remove_space(s):
    return ''.join(ch for ch in s if ch not in string.whitespace)
remove_space("a b\tc\nd")
```
