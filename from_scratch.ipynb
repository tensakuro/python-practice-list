```python
"""
Example demonstrating:
- defining functions
- calling functions
- difference between print and return
"""

# ------------------------------------------------------------
# Function that performs an action (side effect: printing)
# It does NOT return a value, so return type is None
# ------------------------------------------------------------
def print_lyrics() -> None:
    """Print a simple message."""
    print("I'm going to sleep")


# Calling the function
print_lyrics()

# DO NOT do this normally:
# This prints the function output, then prints its return value (None)
print(print_lyrics())


# ------------------------------------------------------------
# Function calling another function (composition of behavior)
# ------------------------------------------------------------
def repeat_lyrics() -> None:
    """Call print_lyrics twice."""
    print_lyrics()
    print_lyrics()


repeat_lyrics()


# ------------------------------------------------------------
# Function that RETURNS a value instead of printing
# ------------------------------------------------------------
def greet(name: str) -> str:
    """
    Build and return a greeting message.

    Parameters:
        name: person's name

    Returns:
        A formatted greeting string.
    """
    return "Hello " + name


# Using the returned value
message = greet("Sohel")
print(message)
```

```python
# -----------------------------------------------------------
# This function prints the given message two times.
# The parameter 'message' receives whatever value is passed
# during the function call.
# -----------------------------------------------------------
def print_twice(message):
    print(message)
    print(message)


# -----------------------------------------------------------
# This function combines two pieces of text and prints
# the combined result once.
#
# first_text  -> first input string
# second_text -> second input string
#
# A local variable 'combined_text' is created inside
# this function. It exists only while the function runs.
# -----------------------------------------------------------
def print_combined(first_text, second_text):

    # Concatenate the two strings with a space between them
    combined_text = first_text + " " + second_text

    # Output the combined text
    print(combined_text)

    # If we want it printed twice instead of once,
    # we can call another function:
    # print_twice(combined_text)


# -----------------------------------------------------------
# Main program (top level / __main__)
# Variables created here are global to this file.
# -----------------------------------------------------------

line1 = "bigg back"
line2 = "baloch bilol"

# Call the function and pass the variables as arguments
print_combined(line1, line2)
```

```python
# ------------------------------------------------------------
# right_justify
# Aligns a string so its last character appears in column 70.
# Instead of printing, the function RETURNS the formatted text
# so it can be reused (printed, stored, or written to a file).
# ------------------------------------------------------------

def right_justify(s: str) -> str:
    # Target display column for the final character
    TARGET_COLUMN = 70

    # Calculate how many spaces must appear before the string
    # Example:
    # length = 5  → spaces = 70 - 5 = 65
    leading_spaces = TARGET_COLUMN - len(s)

    # Prevent negative spacing if the string is longer than 70
    # (in that case, just return the original string)
    leading_spaces = max(0, leading_spaces)

    # Build the final right-aligned string
    formatted_text = " " * leading_spaces + s

    # Send the formatted result back to the caller
    return formatted_text


# Example usage
print(right_justify("sohel"))
```

```python
def print_horizontal():
    for i in range(2):
        print("+",end=" ")
        for j in range(4):
            print("-",end=" ")
    print("+")
def print_vertical():
    for i in range(2):            # two cells
        print("|", end=" ")
        for j in range(4):        # spacing inside the cell
            print(" ", end=" ")
    print("|")
    
def print_grid():
    for i in range(2):            # two rows of boxes
        print_horizontal()

        for j in range(4):        # four interior lines
            print_vertical()

    print_horizontal()  
    
print_grid()# bottom border
```
