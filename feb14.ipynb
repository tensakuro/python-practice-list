```python
def calculate(op,*nums):
    if op=="add":
        s=0
        for x in nums:
            s+=x
        return s
    elif op =="mul":
        product=1
        for x in nums:
            product*=x

        return product
    
print(calculate("add",4,5,6))
print(calculate("mul",4,5,6))
```
