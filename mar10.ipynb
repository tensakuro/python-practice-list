```python
n=5
for r in range(1,n+1):
    for c in range(1,r+1):
        print("*",end=" ")
    print()
```

```python
n=4
for r in range(n):
    for c in range(n):
        print("*",end=" ")
    print()
```

```python
n=5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
