```python
num=int(input("enter number: "))
temp=num

count=0

while temp>0:
    temp//=10
    count+=1 #no of digits

temp=num

total=0

while temp>0:
    digit=temp%10
    total+=digit**count
    temp//=10

print(f"{num} is {'an armstrong number' if total==num else 'not a armstrong'}")
```

```python
num=int(input("enter the number: "))

digits=str(num)
power=len(digits)

total=0

for d in digits:
    total+=int(d) ** power
print(f"{num} is {'an Armstrong number' if total == num else 'not an Armstrong number'}")
```

```python
num=int(input("enter number: "))
power=len(str(num))

total=sum(int(d)**power for d in str(num))

print(f"{num} is {'an Armstrong number' if total == num else 'not an Armstrong number'}")
```
