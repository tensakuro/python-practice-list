```python


first_name=input("enter the firstname: ")
last_name=input("enter the last name: ")
city=input("enter the city: ")
age=int(input("enter the age"))
fav_animal=input("enter the name of fav_animal: ")
fav_number=int(input("enter favourite number: "))


def extract_initials(first_name,last_name):
    return first_name[0].upper()+last_name[0].upper()


def extract_first_n_chars(c:str,n:int)->str:

    # handle invalid cases

    if c=="" or n<=0:
        return ""
    
    res=""
    i=0
    # stop when either limit is reached

    while i<n and i<len(c):
        res=res+c[i]
        i=i+1

    return res

def checking_eligibility(age:int)->bool:
    return age>=18


def combine_inputs(first_name,last_name,city,age,fav_animal,fav_number):
    return [first_name,last_name,city,age,fav_animal,fav_number]

initials=extract_initials(first_name,last_name)

print("initials:",initials)

first_three_char=extract_first_n_chars(city,3)

print(f" first three characters of my favourite city is {first_three_char}" )
checkvote=checking_eligibility(age)
if checkvote:
    print(f"{first_name.title()+' '+last_name.title()} is eligible to vote"+" "+"because age is greater than or equal to 18")

else:
    print(f"{first_name.title()+' '+last_name.title()} is not eligible to vote"+" "+"because age is less than 18")

all_data=combine_inputs(first_name,last_name,city,age,fav_animal,fav_number)
print("combined list of all inputs:",all_data)


    

    


```
