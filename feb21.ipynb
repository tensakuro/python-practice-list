```python
def greet(name: str) -> str:
    """
    Return a greeting message for the given name.

    Example:
        greet("Sohel") -> "Hello, Sohel how are you doing today?"
    """
    message = f"Hello, {name} how are you doing today?"
    return message


if __name__ == "__main__":
    print(greet("Sohel"))
```
