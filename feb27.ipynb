```python
def sum_array(arr):
    if arr is None or len(arr)<3:
        return 0
    
    total=sum(arr)
    smallest=min(arr)
    largest=max(arr)
    return total -smallest-largest

sum_array([6,2,1,8,10])
```

```python
def sum_array(arr):
    if arr == None or len(arr)<3:
        return 0
    
    total=0
    smallest=float('inf')
    largest=float('-inf')
    
    for x in arr:
        total+=x
        if x < smallest:
            smallest=x
        if x > largest:
            largest=x
    
    return total - smallest - largest

sum_array([6,2,1,8,10])
```
