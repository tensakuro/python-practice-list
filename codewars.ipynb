```python
def make_negative( number ):
    if number < 0:
        return number
    if number > 0:
        return (-1) * number
    else:
        return 0
    

make_negative(1)
make_negative(0)
make_negative(-9)
```

```python
def get_duplicates(nums:list)->list:
    freq={}
    duplicates=[]
    for num in nums:
        freq[num]=freq.get(num,0)+1
        if freq[num]==2:
            duplicates.append(num)
    
    return duplicates

get_duplicates([1,2,3,2,4,1,5])
```
