```python
votes = ["Alice", "Bob", "Alice", "Charlie", "Alice", "Bob"]

vote_count={}

for vote in votes:
    if vote in vote_count:
        vote_count[vote]+=1
    else:
        vote_count[vote]=1
        
for name in vote_count:
    print(name,":",vote_count[name])
```
