```python
# ── Sample Data ───────────────────────────────────────────────
orders = [
    {"order_id": 1,  "customer": "Alice",   "product": "Laptop",  "price": 75000, "quantity": 1, "status": "delivered"},
    {"order_id": 2,  "customer": "Bob",     "product": "Phone",   "price": 25000, "quantity": 2, "status": "delivered"},
    {"order_id": 3,  "customer": "Alice",   "product": "Tablet",  "price": 30000, "quantity": 1, "status": "cancelled"},
    {"order_id": 4,  "customer": "Charlie", "product": "Laptop",  "price": 75000, "quantity": 2, "status": "delivered"},
    {"order_id": 5,  "customer": "Bob",     "product": "Phone",   "price": 25000, "quantity": 3, "status": "delivered"},
    {"order_id": 6,  "customer": "Diana",   "product": "Monitor", "price": 15000, "quantity": 2, "status": "cancelled"},
    {"order_id": 7,  "customer": "Charlie", "product": "Laptop",  "price": 75000, "quantity": 1, "status": "delivered"},
    {"order_id": 8,  "customer": "Alice",   "product": "Phone",   "price": 25000, "quantity": 2, "status": "delivered"},
    {"order_id": 9,  "customer": "Diana",   "product": "Tablet",  "price": 30000, "quantity": 1, "status": "delivered"},
    {"order_id": 10, "customer": "Bob",     "product": "Monitor", "price": 15000, "quantity": 3, "status": "delivered"},
]

delivered=[]
cancelled=[]

for order in orders:
    if order["status"]=="delivered":
        delivered.append(order)
    else:
        cancelled.append(order)
print(f"Delivered : {len(delivered)} | Cancelled : {len(cancelled)}") 





```

```python

def get_total_revenue(orders):
    total_revenue=0
    for order in orders:
        if order["status"]=="delivered":
            total_revenue+=(order["price"] * order["quantity"])
    return total_revenue


revenue=get_total_revenue(orders)

print("total revenue:",revenue)
```

```python
def get_most_sold_product(orders):
    product_units={}
    for order in orders:
        if order["status"]=="delivered":
            product=order["product"]
            quantity=order["quantity"]
            
            if product in product_units:
                product_units[product]+=quantity
            else:
                product_units[product]=quantity
                
    most_sold=None
    highest=0
    
    for product in product_units:
        if product_units[product] > highest:
            highest=product_units[product]
            most_sold=product
    return most_sold,highest


product,units=get_most_sold_product(orders)


print("most sold product",product)
print("units sold",units)
    
    
```

```python
def get_top_customer(orders):
    customer_spending={}
    
    for order in orders:
        if order["status"]=="delivered":
            customer=order["customer"]
            order_revenue=order["price"] * order["quantity"]
            
            if customer in customer_spending:
                customer_spending[customer]+=order_revenue
            else:
                customer_spending[customer]=order_revenue
    
    top_customer=None
    highest_spent=0
    
    for customer in customer_spending:
        if customer_spending[customer]>highest_spent:
            highest_spent=customer_spending[customer]
            top_customer=customer
            
    return top_customer,highest_spent

get_top_customer(orders)
```

```python
def get_cancelled_orders(orders):
    cancelled_list=[]
    cancelled_count=0
    revenue_lost=0
    
    for order in orders:
        if order["status"]=="cancelled":
            cancelled_list.append(order)
            cancelled_count+=1
            revenue_lost+=(order["price"]*order["quantity"])
    return cancelled_list,cancelled_count,revenue_lost

get_cancelled_orders(orders)
```
