```python
# ------------------------------------------------------------
# nested_sum(numbers)
# Returns the total of all integers contained inside
# a list of lists.
#
# Example:
# [[1,2,3], [4,5], [6]]  →  21
# ------------------------------------------------------------
def nested_sum(numbers: list[list[int]]) -> int:

    total_sum = 0   # accumulator for the final result

    # Outer loop: visit each inner list
    for sublist in numbers:

        # Inner loop: visit each number inside that list
        for value in sublist:
            total_sum += value

    return total_sum

```

```python
def cumsum(numbers:list[int])->list[int]:
    result=[]
    running_total=0
    for value in numbers:
        running_total+=value
        result.append(running_total)
        
    return result
    
    
cumsum([1,2,3,4])
```

```python
def is_sorted(lst:list)->bool:
    """return true if the list is in ascending order."""
    for i in range(len(lst)-1):
        if lst[i]>lst[i+1]:
            return False
    return True
```

```python
def is_anagram(word1:str,word2:str)->bool:
    """return true if two words are anagrams"""
    word1=word1.lower()
    word2=word2.lower()
    
    return sorted(word1) == sorted(word2)

is_anagram("sohel","lehos")


"""with out sorting"""

def iss_anagram(word1:str,word2:str)->bool:
    if len(word1)!=len(word2):
        return False
    counts={}
    
    for ch in word1:
        counts[ch]=counts.get(ch,0)+1
    for ch in word2:
        if ch not in counts:
            return False
        counts[ch]-=1
        if counts[ch]<0:
            return False
    return True

iss_anagram("sohel","lesos")

def is_anagram_two_maps(word1:str,word2:str)->bool:
    if len(word1)!=len(word2):
        return False
    count1={}
    count2={}
        
    for ch in word1:
        count1[ch]=count1.get(ch,0)+1
    for ch in word2:
        count2[ch]=count2.get(ch,0)+1
            
    return count1==count2
    
is_anagram_two_maps("lehos","sohel")
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
```

```python
"""character frequency counter"""

def histogram(s:str):
    frequency={}
    
    for ch in s:
        frequency[ch]=frequency.get(ch,0)+1
    return frequency


histogram("mississippi")
```

```python
"""2) First non-repeating character"""

def first_unique(s:str):
    freq={}
    for ch in s:
        freq[ch]=freq.get(ch,0)+1
        
    for ch in s:
        if freq[ch]==1:
            return ch
    return None

first_unique("swiss")
```

```python
"""most frequent character"""

def most_frequent(s):
    freq={}
    for ch in s:
        freq[ch]=freq.get(ch,0)+1
    max_char=None
    max_count=0
    
    for ch in freq:
        if freq[ch]>max_count:
            max_count=freq[ch]
            max_char=ch
    return max_char

most_frequent("banana")
```

```python
def reverse_lookup_all(d, v):
    result = []
    for k, value in d.items():
        if value == v:
            result.append(k)

    if not result:
        raise LookupError("value not found")

    return result


d = {'a':1,'b':2,'c':1}

print(reverse_lookup_all(d,1))
```

```python
"""1) Find all students with a given grade"""

grades = {
    "sohel": "A",
    "rahul": "B",
    "anita": "A",
    "kiran": "C"
}
"""Return all students who got "A"."""

def students_with_grade(grades,target):
    result=[]
    for name,grade in grades.items():
        if grade==target:
            result.append(name)
    return result

print(students_with_grade(grades,"A"))
```

```python
""" invert a dictionary"""

def invert_dict(d):
    inverse={}
    for key in d:
        val=d[key]
        if val not in inverse:
            inverse[val]=[key]
        else:
            inverse[val].append(key)
    return inverse

hist = {'a':1,'p':1,'r':2,'t':1,'o':1}

invert_dict(hist)
```
