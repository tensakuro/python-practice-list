```python
num=int(input("enter the number: "))

o=num
total=0

while num>0:
    num,dig=divmod(num,10)
    i=1
    fact=1
    while i<=dig:
        fact*=i
        i=i+1
    total+=fact

if total==o:
    print(f"{o}is a strong")
else:
    print(f"{o}not a strong")
```
