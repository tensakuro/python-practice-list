```python
def rows_brute(n):
    row=1
    count=0
    
    while n>=row:
        n-=row
        count+=1
        row+=1
        
    return count

print(rows_brute(8))
```

```python
def is_perfect_square(n):
    if n < 0:
        return False
    i=0
    while i*i <=n:
        if i * i ==n:
            return True
        i+=1
        
    return False

is_perfect_square(81)
            
        
            
```
