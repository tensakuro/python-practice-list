```python
def rows_brute(n):
    row=1
    count=0
    
    while n>=row:
        n-=row
        count+=1
        row+=1
        
    return count

print(rows_brute(8))
```

```python
# sliding window ,maximum size of subarray of window size =3

arr =[2,1,5,1,3,2]

k=3



def max_size_sub(arr,k):
    
    n=len(arr)
    
    if k > n:
        return None
    win_sum = 0
    max_sum = float('-inf')
    for i in range(k):
        win_sum += arr[i]
    max_sum = win_sum
    
    for i in range(k,n):
        win_sum = win_sum - arr[i-k] + arr[i]
        
        if win_sum > max_sum:
            max_sum =  win_sum
            start = i - k + 1
            end = i + 1
            
    
    return (max_sum,arr[start:end])

max_size_sub(arr,k)
    
            
```

```python
def match_sub_array(arr,k, target_comp):
    n = len(arr)
    win_sum = 0
    
    if k > n:
        return None
    
    for i in range(k):
        win_sum =  win_sum + arr[i]
    
    if win_sum > target_comp:
        return arr[0:k]
    
    
    for i in range(k,n):
        win_sum = win_sum - arr[i-k] + arr[i]
        
        if win_sum > target_comp:
            start = i -(k-1)
            end = i + 1
            
            return arr[start : end]
    return None
    
    

arr = [2, 1, 5, 1, 3, 2]
k = 3
target_comp = 7

match_sub_array(arr,k,target_comp)
```

```python
def max_sub_bruteforce(arr):
    n = len(arr)
    best = float('-inf')
    
    for i in range(n):
        s = 0
        for j in range(i,n):
            s+=arr[j]
            if s > best:
                best=s
    return best
```

```python
def maxsubarray(arr):
    curr=arr[0]
    best = arr[0]
    
    for x in arr[1:]:
        curr=max(x,x+curr)
        best =max(best,curr)
    return best

```
