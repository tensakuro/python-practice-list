```python
n=int(input("enter a number"))

if n<=1:
    print(" not a prime")
else:
    for i in range(2,n):
        if n%i==0:
            print("not  a prime")
            break
    else:
        print("prime")
    
    
    
        
            
    
```

```python
n=int(input("enter a num"))

factors=[]

for i in range(1,n+1):
    if n%i==0:
        factors.append(i)
        
print(factors)
        

if n<=1:
    print("not a prime")
elif len(factors)==2:
    print("prime")
else:
    print("not a prime")

    
    
        
```

```python
num=int(input("enter a number"))

while num >=10:
    sum=0
    while num!=0:
        sum+=num%10
        num//=10
    num=sum
    
print(num)


```
