```python
#find the second largest number
nums=[4,1,9,3,9,7]
def second_largest(nums):
    unique=list(set(nums))
    unique.sort()
    return unique[-2]

second_largest(nums)
```

```python
def second_largest(arr: list):
    if len(arr) < 2:
        return None

    large = float('-inf')
    second_large = float('-inf')

    for num in arr:

        if num > large:
            second_large = large
            large = num

        elif num > second_large and num != large:
            second_large = num

    if second_large == float('-inf'):
        return None

    return second_large


nums = [4, 1, 9, 3, 9, 7]
print(second_largest(nums))
```
