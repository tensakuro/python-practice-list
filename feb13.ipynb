```python
def is_even(n):
    return n%2==0

n=int(input("enter the number"))

if is_even(n):
    print(f"{n} is a even number")
else:
    print(f"{n} is a odd number")
```

```python
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
```
