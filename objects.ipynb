```python
# Task 1: Basic Inheritance
# Create a Parent class Person with method eat().
# Create a Child class Student that inherits Person and has its own method study().
# Create an object of Student and call both methods.

class Parent:
    def __init__(self,name):
        self.name=name
        
    def eat(self):
        print(f"{self.name}  is eating")
        
    def __str__(self):
        return f"{self.name} is introduced"
class Child(Parent):
    def study(self):
        print(f"{self.name} is studying")
        
a=Child("sohel")


print(a)
# print(a.eat())
# print(a.study())

a.eat()
a.study()


```

```python
# Task 2: Inherited Method Usage
# Create a parent class Vehicle with method start().
# Create a child class Car that inherits Vehicle without redefining start().
# Call start() using Car object

class Vehicle:
    def __init__(self,name):
        self.name = name
        
    def start(self):
        return f"{self.name} is starting soon"
    
    def __str__(self):
        return f"name of the vehicle is {self.name}"
    
class Car(Vehicle):
    def rise(self):
        return f"{self.name} is rising fast"

porche = Car("sedan")


porche

print(porche)


porche.start()


porche.rise()
    
```

```python
# Task 3: Specialized (Specific) Method
# Create a parent class Employee with method work().
# Create a child class Developer that inherits Employee and adds specialized method write_code().
# Call both methods using child object

class Employee:
    def __init__(self,name):
        self.name = name
        
    def work(self):
        return f"{self.name} is currently working"
    
    def __str__(self):
        return f"name of them developer is {self.name}"
    
class Developer(Employee):
    def write_code(self):
        return f"the {self.name} is writing the code"
    
dev = Developer("rajesh")

ded= dev.work()


des=dev.write_code()

print(ded)

print(des)
```

```python
# Task 4: Method Overriding
# Create a parent class Animal with method sound() that prints "Animal makes sound".
# Create a child class Dog that overrides sound() to print "Dog barks".
# Call sound() using Dog object.

class Animal:
    def __init__(self,name):
        self.name=name
    def sound(self):
        return "Animal is making sound"
class Child(Animal):
    def sound(self):
        return "dog barks"

    def sound_cat(self):
        return f"{self.name} is meowing"
    
cat = Child("ringa")

cat.sound()

cat.sound_cat()
```

```python
# Task 5: Simple Overriding
# Create a parent class Shape with method draw() that prints "Drawing shape".
# Create a child class Circle that overrides draw() to print "Drawing circle".
# Call draw() using Circle object.

class Shape:
    def __init__(self,name):
        self.name =  name
        
    def draw(self):
        return "Drawing shape"
    
class Circle(Shape):
    def draw(self):
        return "Drawing circle"
    
cir = Circle("circle")

cir.draw()
```
