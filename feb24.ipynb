```python
# reverse a list in place
# two pointer swap

def reverse_list(arr:list)->list:
    left=0
    right=len(arr)-1
    while left < right:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1
    return arr

nums=[1,2,3,4,5]
print(reverse_list(nums))
```

```python
# kth largest using sorted function

def kth_largest(nums,k):
    if k<=0 or k>len(nums):
        return None
    nums_sorted=sorted(nums,reverse=True)
    return nums_sorted[k-1]
nums=[7,2,9,4,1,8]
kth_largest(nums,2)
```

```python
import random

def kth_largest(nums, k):
    if k <= 0 or k > len(nums):
        return None
    
    target = len(nums) - k

    def quickselect(left, right):
        pivot = nums[random.randint(left, right)]
        
        i, j = left, right
        
        while i <= j:
            while nums[i] < pivot:
                i += 1
            while nums[j] > pivot:
                j -= 1
            if i <= j:
                nums[i], nums[j] = nums[j], nums[i]
                i += 1
                j -= 1

        # IMPORTANT: this part must be OUTSIDE the loop
        if target <= j:
            return quickselect(left, j)
        elif target >= i:
            return quickselect(i, right)
        else:
            return nums[target]

    return quickselect(0, len(nums)-1)
```

```python
# find max in list
def find_max(arr:list):
    if len(arr)==0:
        return None
    large=arr[0]
    for num in arr:
        if num>large:
            large=num
    return large

nums=[7,2,9,4,1,8]

print(find_max(nums))
```

```python
# rank and how many numbers large than our target

def count_greater(arr:list,x:int)->int:
    count=0
    for num in arr:
        if num>x:
            count+=1
    return count
```

```python
# Write a function that returns how many times each character appears.

from collections import Counter
def char_count(s:str):
    return Counter(s)

print(char_count("banana"))
```

```python
# Q2. Find most frequent number

from collections import Counter

def most_freq(arr):
    return Counter(arr).most_common(1)

most_freq([1,2,2,3,3,3,4])
```

```python
# return most frequest element
from collections import Counter
def most_element(arr):
    return Counter(arr).most_common(1)[0][0]

most_element([5,5,1,2,2,2,3])
```

```python
# return top-k elements

from collections import Counter

def top_k(arr,k):
    return [x for x,_ in Counter(arr).most_common(2)]

top_k([1,1,1,2,2,3,3,3,4],2)
```
