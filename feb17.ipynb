```python
def factorial(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f

def is_strong(num):
    temp=num
    s=0
    while temp>0:
        digit=temp%10
        s+=factorial(digit)
        temp//=10

    return s==num



is_strong(145)

def is_neon(num):
    sq=num*num
    s=0
    while sq>0:
        digit=sq%10
        s+=digit
        sq//=10
    return s==num

is_neon(9)
```

```python
def wallet():
    balance = 1000

    def deposit(amount):
        nonlocal balance
        balance += amount
        print("After deposit:", balance)

    def withdraw(amount):
        nonlocal balance
        balance -= amount
        print("After withdrawal:", balance)

    return deposit, withdraw


add_money, take_money = wallet()

add_money(500)
take_money(200)
take_money(300)
```

```python
def recu(n):
    if n == 0:
        return
    print(n)
    recu(n-1)

recu(10)
```
