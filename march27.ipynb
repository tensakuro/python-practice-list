```python
# find max number in a list



def maxinlist(lst:list):
    max_num=lst[0]
    # max_num=float('-inf')
    for num in lst:
        if num > max_num:
            max_num=num
    return max_num


max_number=maxinlist([1,4,6,2,0,8])

print(max_number)
```
